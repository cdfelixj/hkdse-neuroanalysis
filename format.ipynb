```python
import pandas as pd

df = pd.read_csv("2024_HKDSE_results_statistics_table3g_en.csv")


df.rename(columns={'Description': 'description',
                   'Type': 'type',
                   'Physical Disabilities': 'physical_disabilities',
                   'Visual Disabilities': 'visual_disabilities',
                   'Aural Disabilities': 'aural_disabilities',
                   'Oral Disabilities': 'oral_disabilities',
                   'Specific Learning Disabilities': 'specific_learning_disabilities',
                   'Autism Spectrum Disorder': 'autism_spectrum_disorder',
                   'Attention Deficit/Hyperactivity Disorder': 'adhd',
                   'Mental Derangement': 'mental_derangement',
                   'Other Disabilities': 'other_disabilities',
                   'Total': 'total'}, inplace=True)


df['description'] = df['description'].str.replace('No. ', '', regex=False)


df = df[df['description'] != 'sat']


df.set_index('description', inplace=True)


cols_to_numeric = df.columns.drop('type')
df[cols_to_numeric] = df[cols_to_numeric].apply(pd.to_numeric, errors='coerce')

df_levels = df.iloc[0:4]  
df_core = df.iloc[4:]  

df_levels.to_csv('HKDSE_levels_data.csv')
df_core.to_csv('HKDSE_core_subjects_data.csv')

print("DataFrame split and exported to HKDSE_levels_data.csv and HKDSE_core_subjects_data.csv")
```
